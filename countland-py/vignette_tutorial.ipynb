This tutorial demonstrates the major functions of `countland` by applying them to a Gold standard single-cell RNA sequencing dataset from [Freytag _et al_ (2018)](10.12688/f1000research.15809.2).

`countland` is a **barebones** set of functions for applying a restricted linear algebra to the analysis of count-based data. As such, there are many opportunities for further optimization that may prove useful in the anlaysis of your own data. We provide the source code freely available at https://github.com/shchurch/countland and encourage users and developers to fork the code for their own purposes.

## GETTING STARTED

The following packages are required to complete the tutorial.

```python
import countland
import scanpy as sc
import numpy as np

import matplotlib
matplotlib.pyplot.style.use('classic')
%config InlineBackend.figure_formats = ['svg']
%matplotlib inline
```

## GET THE DATA

We have used the Gold standard dataset provided by [Freytag _et al_ (2018)](10.12688/f1000research.15809.2). This data consists of ~1000 cells that have ground truth labels corresponding to three human lung cell lines.

`countland` accepts an AnnData object (as does `scanpy`). 

```python
adata = sc.read_10x_mtx(
    "../data/Gold_Freytag2018",  # the directory with the `.mtx` file
    var_names='gene_symbols',    # use gene symbols for the variable names (variables-axis index)
    cache=True)                  # write a cache file for faster subsequent reading

adata.var_names_make_unique()    # this is unnecessary if using `var_names='gene_ids'` in `sc.read_10x_mtx`
```

## INITIALIZE COUNTLAND OBJECT

First we construct an instance of a `countland` object, loading it with the data from the AnnData object.

```python
C = countland.countland(adata)
print(C)
```

The count matrix is stored in `C.counts`

```python
C.counts
```

Note that most counts are zero for scRNA-seq data.

## EXPLORE DATA CHARACTERISTICS

Here we explore basic characteristics of the data. In the following dataframe, each row is a cell and each column is a different summary statistic of counts in these cells. These include:

- maximum number of counts observed for any feature in this cell
- total counts across all features
- number of features with observed counts
- number of features with counts above 1, or 10
- number of unique count values across all features
- count index = number of _n_ genes with _n_ counts (think _h_ index, but for counts)

It can also be helpful to see how many counts are derived from certain genes, such as mitochondrial genes. Here we use a regular expression to match a gene by name, and the counts for this feature are added as the `feature_match_counts` column.

```python
C.ScoreCells(gene_string="140988$")
C.cell_scores
```

## FILTER DATA

Standard workflows often filter data based on various thresholds, such as the number of unique features per cell. 

```python
filter_cells = np.array(C.cell_scores['n_features'] < 8500) # cells with fewer than 8,500 unique genes observed
C.SubsetCells(filter_cells,remove_empty=False)
```

```python
filter_genes = (np.sum(C.counts,0) > 0) # genes that were observed
C.SubsetGenes(filter_genes,remove_empty=False)

# note the above could be accomplished by running
#     C.SubsetCells(filter_cells,remove_empty=True)
```

With `countland`, such data filtering may not be necessary or helpful. The original count matrix can be restored at any time.

```python
C.RestoreCounts()
```

## SUBSAMPLE GENES

Cells are not sequenced to standard sequencing depth. This is sometimes a problem for downstream comparisons.  

You can create an alternative count matrix where cells have an equal number of counts using `C.Subsample()`. The number of counts must not be larger than the minimum number per cell.

This matrix is stored in `C.subsample`

```python
np.random.seed(84095) # choose a random seed for reproducibility
C.Subsample(n_counts = 1000)
```

## COMPARE GENES BY COUNTS

Using the subsampled count matrix, we can make comparisons of gene expression across cells using several count-based assessments. These include:
- maximum number of counts observed across cells
- total counts
- number of cells with observed counts
- number of cells with counts above 1, or 10
- number of unique count values
- count index = number of _n_ cells with _n_ counts

```python
C.ScoreGenes(subsample=True)
C.gene_scores.sort_values(by="unique_count_values",ascending=False)
```

We can also visualize counts across genes with a stripplot

```python
gene_indices = C.gene_scores.sort_values(by="unique_count_values",ascending=False).head(10).index
C.PlotGeneCounts(gene_indices)
```

## COMPARE CELLS AND CLUSTER

The dot product (also known as the inner product) indicates the alignment between vectors. It is 0 when the vectors are perpendicular, and increases in length as the factors get longer and/or more aligned.

In our analyses, it tells us how similar two cells are based on their proportions of counts, scaled by the total counts in each cell. A dot product of 0 indicates orthogonal cell vectors (no shared counts), larger values indicate more aligned cell vectors.

```python
C.Dot()
```

Cell populations can be compared and distinguished by embedding and clustering the matrix of pairwise dot products (contained in `C.dots`). This matrix is an unbounded affinity matrix. It is symmetric, and contains only natural numbers (0 and positive integers). Spectral embedding clustering is appropriate for this type of matrix.

First, we apply spectral embedding on our data.

```python
C.Embed()
```

We then use an eigengap heuristic is can help decide the optimal number of clusters. Note that this is only a guideline. According to this heuristic, the optimal number of clusters is _k_ where the difference in eigenvalues $ | \lambda_{k+1} - \lambda_{k} | $ is largest.

```python
C.PlotEigengap()
```

For many datasets, you may want to consider other factors, e.g. choosing a minimum number of clusters, whether or not the eigengap reflects this.

Here we have chosen 3 as the optimal number of clusters.

```python
C.Cluster(n_clusters=3)
```

### VISUALIZING WITH SPECTRAL EMBEDDING OF DOT PRODUCTS

We can now visualize clusters using spectral embedding.

```python
C.PlotEmbedding()
```

## IDENTIFY MARKER GENES

The choice of an ideal marker gene for a given cluster will depend on downstream applications. For example, the ideal marker gene might be defined as the gene detected in all of the cells in a given cluster and none of the cells in other clusters.

Under this definition, the top marker gene for each cluster can be identified by counting and comparing the number of non-zero cells.

```python
C.RankMarkerGenes(method='prop-zero')
gdf = C.marker_genes.loc[(C.marker_genes['cluster_label'] == 1)]
gdf
```

```python
gene_index = gdf.loc[(gdf['cluster_label'] == 1) & (gdf['rank'] == 1)]['gene index'].values
C.PlotMarker(gene_index)
```

Alternatively, the top marker gene might be defined as the gene with the highest differential expression. In this case, the top marker for each cluster can be identified by ranking genes using the Wilcoxon rank-sum statistic.

**Note this can take some time.**

```python
C.RankMarkerGenes(method='rank-sums')
gdf = C.marker_genes.loc[(C.marker_genes['cluster_label'] == 1)]
gdf
```

```python
gene_index = gdf.loc[(gdf['cluster_label'] == 1) & (gdf['rank'] == 1)]['gene index'].values 
C.PlotMarker(gene_index)
```
